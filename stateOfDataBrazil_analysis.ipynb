Este bloco de código converte o arquivo para UTF-8 em seguida cria outro arquivo com dados convertidos e salva no formato **.csv**

```python
import pandas as pd


State_of_data_BR_2023 = pd.read_excel("/content/State_of_data_BR_2023.xlsx")

tabela = pd.DataFrame(State_of_data_BR_2023)

tabela.columns = tabela.columns.str.encode('latin1').str.decode('utf-8')

# Converte os dados para UTF-8
def converte_texto_para_utf8(celula):
    if isinstance(celula, str):
        try:
            return celula.encode('latin1').decode('utf-8')
        except (UnicodeEncodeError, UnicodeDecodeError):
            # Se ocorrer um erro, retorna o valor original
            return celula
    return celula


for column in tabela.columns:
    tabela[column] = tabela[column].apply(converte_texto_para_utf8)


# Salva uma cópia desses dados convertidos
tabela.to_csv("tabela_convertida.csv", encoding='utf-8', index=False)

```

Bloco de código que retira todo os caracteres especiais do arquivo que foi convertido para UTF-8.

```python
import pandas as pd


tabela_convertida = pd.read_csv("/content/tabela_convertida.csv")

df = pd.DataFrame(tabela_convertida)


# Dicionário de substituições
substituicoes = {
    'ç': 'c', 'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
    'ã': 'a', 'õ': 'o', 'â': 'a', 'ê': 'e', 'ô': 'o', 'à': 'a',
    'ü': 'u', 'Ç': 'C', 'Â': 'A', 'Ê': 'E', 'Ô': 'O',
    'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U'
}


# Criar uma tabela de tradução
tabela_traducao = str.maketrans(substituicoes)

# Aplicar a tradução a todas as células do DataFrame
df = df.applymap(lambda celula: celula.translate(tabela_traducao) if isinstance(celula, str) else celula)


# Salva uma cópia desses dados convertidos
df.to_csv("stateOfDataBR2023_sem_caracteres_especiais.csv", encoding='utf-8', index=False)
```

### Realizado últimos tratamentos nos títulos das colunas onde foram removidos: espacos em branco e substituído por underline(_), virgula por travessão, aspas simles e removido caracteres especiais.

```python
import pandas as pd


stateOfDataBR2023 = pd.read_csv("/content/stateOfDataBR2023_sem_caracteres_especiais.csv")

df = pd.DataFrame(stateOfDataBR2023)

# Dicionário de substituições
substituicoes = {
    'ç': 'c', 'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u',
    'ã': 'a', 'õ': 'o', 'â': 'a', 'ê': 'e', 'ô': 'o', 'à': 'a',
    'ü': 'u', 'Ç': 'C', 'Â': 'A', 'Ê': 'E', 'Ô': 'O',
    'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U'
}

# Criar uma tabela de tradução
tabela_traducao = str.maketrans(substituicoes)

# Aplicar a substituição dos caracteres especiais por normal no DataFrame
df.rename(columns=lambda coluna: coluna.translate(tabela_traducao), inplace=True)



# Remove aspas simples e virgulas presente na coluna.
df.columns = df.columns.str.replace("'", "").str.replace(",", "-").str.replace(" ", "_")


df.to_csv("stateOfDataBR2023_tratado.csv", encoding='utf-8', index=False)

```

```python
import pandas as pd

stateOfDataBR2023 = pd.read_csv("/content/stateOfDataBR2023_tratado.csv")

df = pd.DataFrame(stateOfDataBR2023)

# Exibe nome das colunas
for i in range(len(df.columns)):
    print(f'{i}: {df.columns[i]}')
```

#### Este bloco aplica filtros que atendem aos critérios específicos, mantendo apenas as colunas de interesse. Depois agrupa por nível, gênero, soma os totais e por fim cria uma nova coluna com os valores percentuais de cada gênero.

#### Gera gráfico

# Nova seção

```python
def calcular_valor_coluna(merged):
  merged['total_percentual'] = (merged['total'] / merged['total_nivel']) * 100
  return merged



#   O código cria um novo DataFrame chamado 'consulta' que contém apenas as linhas que atendem aos critérios
#  de filtragem especificados, mantendo apenas as colunas de interesse.
consulta = df[
    # (df['(P1_b_-_Genero)'] == 'Feminino') &
    df['(P1_b_-_Genero)'].isin(['Feminino', 'Masculino', 'Prefiro nao informar', 'Outro']) &
    (df['(P2_g_-_Nivel)'].notna()) &
    (df['(P2_a_-_Qual_sua_situacao_atual_de_trabalho?)'].ne('Desempregado, buscando recolocacao')) &
    (df['(P2_a_-_Qual_sua_situacao_atual_de_trabalho?)'].ne('Desempregado e nao estou buscando recolocacao')) &
    (df['(P2_a_-_Qual_sua_situacao_atual_de_trabalho?)'].ne('Somente Estudante (graduacao)')) &
    (df['(P2_a_-_Qual_sua_situacao_atual_de_trabalho?)'].ne('Prefiro nao informar'))
    ][
        ['(P1_b_-_Genero)',
         '(P2_g_-_Nivel)',
         '(P2_a_-_Qual_sua_situacao_atual_de_trabalho?)'
     ]]

# Agrupando por nível e gênero e somando os totais
genero_por_cargo = consulta.groupby('(P2_g_-_Nivel)')['(P1_b_-_Genero)'].value_counts().reset_index(name='total')


# Calculando o total por nível
totais_por_nivel = genero_por_cargo.groupby('(P2_g_-_Nivel)')['total'].sum().reset_index()
totais_por_nivel.rename(columns={'total': 'total_nivel'}, inplace=True)

# Juntando os totais por nível com o DataFrame agrupado
merged = pd.merge(genero_por_cargo, totais_por_nivel, on='(P2_g_-_Nivel)')

# Calculando a porcentagem
# merged['percentual'] = (merged['total'] / merged['total_nivel']) * 100
calcular_valor_coluna(merged)



# Exibindo o resultado final
print(merged[['(P2_g_-_Nivel)', '(P1_b_-_Genero)', 'total', 'total_percentual']])


```

```python
from inspect import stack
import plotly.express as px
grafico = px.bar(merged, x='(P2_g_-_Nivel)', y='total', color='(P1_b_-_Genero)',  title='Gênero por Cargo', barmode='stack')
grafico.show()
```

```python
# Função para calcular a porcentagem de cada valor por nível de cargo
def calcular_valor_coluna(merged):
    merged['total_percentual'] = (merged['total'] / merged['total_nivel']) * 100
    return merged

# Criando o DataFrame 'consulta' filtrando apenas as colunas de interesse
consulta = df[
    (df['P2_g_-_Nivel'].notna()) &  # Nível do cargo
    (df['P8_-_Onde_voce_reside_atualmente?'].notna())  # Local de residência
]

# Agrupando por nível de cargo e local de residência (região) e somando os totais
residencia_por_cargo = consulta.groupby(['P2_g_-_Nivel', 'P8_-_Onde_voce_reside_atualmente?'])\
    .size().reset_index(name='total')

# Calculando o total por nível de cargo
totais_por_nivel = residencia_por_cargo.groupby('P2_g_-_Nivel')['total'].sum().reset_index()
totais_por_nivel.rename(columns={'total': 'total_nivel'}, inplace=True)

# Juntando os totais por nível com o DataFrame agrupado
merged = pd.merge(residencia_por_cargo, totais_por_nivel, on='P2_g_-_Nivel')

# Calculando a porcentagem
merged['percentual'] = (merged['total'] / merged['total_nivel']) * 100
calcular_valor_coluna(merged)

# Exibindo o resultado final
print(merged[['P2_g_-_Nivel', 'P8_-_Onde_voce_reside_atualmente?', 'total', 'total_percentual']])
```
